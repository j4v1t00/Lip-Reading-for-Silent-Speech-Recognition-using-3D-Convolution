```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image,ImageFilter
```

```python
path = 'img_mouth_lipread_resize_demo/IMG_000060.jpg'
```

```python
path = 'img_lipread_mp4_demo/IMG_000043.jpg'
```

```python
image = cv2.imread(path)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
blurred = cv2.GaussianBlur(gray, (3, 3), 0)
```

```python
edged = cv2.Canny(blurred, 10, 100)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))
```

```python
img = Image.open(r'img_lipread_mp4_demo/IMG_000043.jpg')
```

```python
nuevaimg = img.filter(ImageFilter.FIND_EDGES)  #Filtro de FIND_EDGES
plt.imshow(nuevaimg)
```

```python
nuevaimg = img.filter(ImageFilter.EMBOSS)  #Filtro de EMBOSS
plt.imshow(nuevaimg)
```

```python
nuevaimg = img.filter(ImageFilter.CONTOUR)  #Filtro de CONTOUR
plt.imshow(nuevaimg)
```
